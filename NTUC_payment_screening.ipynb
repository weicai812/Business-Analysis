```python
# prompt: make the original amount and credit without the thousand comma (100000.00) then make the amount in 2 decimal place, campare the original amount and credit amount then show the matched if both amount is the same, if not same then show unmatched

import pandas as pd

# Assuming the files are in your current working directory or you provide the full path
file1 = pd.read_csv('ND005.csv')
file2 = pd.read_csv('RA_5600279628.csv')

# Filter file2 for rows where 'DOC TYPE' is 'RE'
filtered_file2 = file2[file2['DOC TYPE'] == 'RE']

# Merge the two dataframes based on the document number
merged_df = pd.merge(file1, filtered_file2, left_on='Document No.', right_on='DOCUMENT NUMBER', how='inner')

# Remove thousands separators and convert to numeric
merged_df['Original Amount'] = merged_df['Original Amount'].astype(str).str.replace(',', '').astype(float)
merged_df['CREDIT'] = merged_df['CREDIT'].astype(str).str.replace(',', '').astype(float)

# Format 'Original Amount' to two decimal places
merged_df['Original Amount'] = merged_df['Original Amount'].round(2)


# Compare the amounts and create a new column 'Status'
merged_df['Status'] = 'Matched'
merged_df.loc[merged_df['Original Amount'] != merged_df['CREDIT'], 'Status'] = 'Unmatched'

# Select the desired columns
result_df = merged_df[['Document No.', 'Original Amount', 'CREDIT', 'DOCUMENT NUMBER', 'Status']]

# Display the result
result_df
```

```python
# prompt: in file 2, if the DOC TYPE = "RS" then the DOCUMENT NUMBER find the External Document No. in file 1 (the external document no. please scan for first 8 digit) , print the DOCUMENT NUMBER result (found or unfound)

import pandas as pd

# Assuming the files are in your current working directory or you provide the full path
file1 = pd.read_csv('ND005.csv')
file2 = pd.read_csv('RA_5600279628.csv')

# Filter file2 for rows where 'DOC TYPE' is 'RS'
filtered_file2 = file2[file2['DOC TYPE'] == 'RS']

# Function to extract the first 8 digits
def extract_first_eight_digits(doc_number):
    try:
      return str(doc_number)[:8]
    except:
      return ''


# Apply the function to the 'External Document No.' column in file1
file1['first_8_digits'] = file1['External Document No.'].apply(extract_first_eight_digits)


# Create an empty list to store the results
results = []

# Iterate through the filtered file2
for index, row in filtered_file2.iterrows():
    document_number = row['DOCUMENT NUMBER']
    external_doc_no_match = file1[file1['first_8_digits'] == str(document_number)[:8]]

    if not external_doc_no_match.empty:
        results.append({'DOCUMENT NUMBER': document_number, 'Result': 'Found'})
    else:
        results.append({'DOCUMENT NUMBER': document_number, 'Result': 'Unfound'})

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Print or save the results DataFrame
results_df
# results_df.to_csv('document_number_results.csv', index=False)
```
